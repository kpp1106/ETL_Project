# Web Scraping Yahoo Finance web page to collect Stock Market Data

```python
import pandas as pd
```

# Defining a variable to store the URL for Yahoo Finance: https://finance.yahoo.com/

```python
url = "https://en.wikipedia.org/wiki/List_of_S%26P_500_companies"
df = pd.read_html(url)
df[0].head()
```

```python
import yfinance as yf

def get_price(ticker):
    return yf.Ticker(ticker)

df_dict = {}

for sym in df[0].Symbol.tolist():
    df_dict[sym] = pd.DataFrame(get_price(sym).history())

# df_ticker_price = df[0].Symbol.apply(lambda yf.Ticker(x).: x)

# df_ticker_price
df_dict[sym].head()
```
